```python
### Local RAG - indexing some search results and then using the index to retrieve the search results again. 
### And basing the generation on the retrieved results

import os 
import pandas as pd 
import numpy as np
```

```python
file_path = './output/searchresults.csv'
df = pd.read_csv(file_path, lineterminator='\n')
```

```python
from sentence_transformers import SentenceTransformer
ST = SentenceTransformer("mixedbread-ai/mxbai-embed-large-v1")
```

```python
df_embedded = df.iloc[0:20] # Select just 20 papers for developing the pipeline 

df_embedded
```

```python

df_embedded['embedding'] = df_embedded.apply(lambda row: ST.encode(row["PaperTitle"] + " " + row["Abstract"]), axis=1)



df_embedded
```

```python
import chromadb
from chromadb.utils import embedding_functions

CHROMA_DATA_PATH = "chroma_data2/"
EMBED_MODEL = "mixedbread-ai/mxbai-embed-large-v1" #"all-MiniLM-L6-v2"
COLLECTION_NAME = "demo"

client = chromadb.PersistentClient(path=CHROMA_DATA_PATH)

embedding_func = embedding_functions.SentenceTransformerEmbeddingFunction(model_name=EMBED_MODEL)

collection = client.create_collection(
     name=COLLECTION_NAME,
     embedding_function=embedding_func,
     metadata={"hnsw:space": "cosine"},
)
```

```python
collection.add(
     documents=[row["PaperTitle"] + " " + row["Abstract"] for i,row in df_embedded.iterrows()],
     ids=[f"id{i}" for i in range(len(df_embedded))],
     metadatas=[{"authors": row['Authors'],"language":row['Lang'],"year":row['PubYear'],"date":row['PubDate']} for i,row in df_embedded.iterrows()]
)
```

```python
import itertools 

query_results = collection.query(
    query_texts=["Which antidepressant is the most effective with the least side effects?"],
    n_results=3,
)

rno = 0 # We only asked one question. This is the answer index. 

for i,rid in enumerate(query_results['ids'][rno]):
    print(rid,query_results['documents'][rno][i],query_results['distances'][rno][i])
```

```python
# A language model to answer questions based on the retrieval results: 

from mlx_lm import load, generate

model, tokenizer = load("mlx-community/Llama-3.2-3B-Instruct")  # Only for MacOS

systemPrompt = "You are a helpful systematic reviewing assistant"

def generateFromPrompt(userPrompt): 

    if hasattr(tokenizer, "apply_chat_template") and tokenizer.chat_template is not None:
        messages = [ {"role": "system", "content": systemPrompt},
              {"role": "user", "content": userPrompt}, ]
        prompt = tokenizer.apply_chat_template(
            messages, tokenize=False, add_generation_prompt=True
        )
        response = generate(model, tokenizer, prompt=prompt, verbose=False)
        return(response)
```

```python
generateFromPrompt("Hello!")
```

```python

prompt = "Please answer the following question using the following paper title and abstract."
query = "What is the role of emotions in psychotherapy?"

query_results = collection.query(
    query_texts=[query],
    n_results=1,
)

title_and_abst = ",".join(query_results["documents"][0])

answer = generateFromPrompt(prompt + query + title_and_abst)

print("Retrieved from ",query_results["ids"][0],"Answer:",answer,"\n\nTitle and abstract:",query_results['documents'][0])
```
